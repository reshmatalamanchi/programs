# Version 2  Keyboard Input



```
try:
    x = int(input("Enter first number: "))
    y = int(input("Enter second number: "))

    result = x + y
    print("Addition using keyboard input:", result)

except ValueError:
    print("Please enter only numbers.")
```

# Version 3  Read Single Input from File



```
with open("single_input.txt", "w") as f:
    f.write("5")

try:
    with open("single_input.txt", "r") as f:
        num = int(f.read())

    print("Number from file:", num)
    print("Result after multiplying by 2:", num * 2)

except FileNotFoundError:
    print("File not found.")

except ValueError:
    print("Invalid data in file.")
```
